```python
#import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#import dataset
df = pd.read_csv("Comcast_telecom_complaints_data.csv")
```

```python
df.head()
```

```python
df["date_index"] = df["Date_month_year"] + " " + df["Time"]
df["date_index"] = pd.to_datetime(df["date_index"])
df["Date_month_year"] = pd.to_datetime(df["Date_month_year"])
df = df.set_index(df["date_index"])

```

```python
#Trend chart for the number of complaints at monthly granularity levels
df.groupby(pd.Grouper(freq="M")).size().plot()
```

```python
#Trend chart for the number of complaints at daily granularity levels
df['Day of Month'] = pd.to_datetime(df["Date"])
df = df.set_index(df["Day of Month"])
df.groupby(pd.Grouper(freq="D")).size().plot()
```

```python
#Provide a table with the frequency of complaint types
df_type = df["Customer Complaint"].value_counts()
df_type.head(25)
```

```python
#since python is case sensistive so Comcast and comcast are shown different, so changing to Upper case completely
df_type = df['Customer Complaint'].str.upper().value_counts()
df_type.head(25)
```

```python
#Create a new categorical variable with value as Open and Closed.
#Open & Pending is to be categorized as Open and Closed & Solved is to be categorized as Closed
df["newStatus"] = ["Open" if Status=="Open" or Status=="Pending" else "Closed" for Status in df["Status"]]
```

```python
df_status = df.groupby('State').newStatus.value_counts().unstack()
#df_status = df_complaints.groupby(['State','newStatus'])['COUNT'].sum().unstack()
```

```python
df_status.head()
```

```python
plt.figure(figsize=(200,100)) 
plt.rcParams['figure.dpi'] = 200
df_status.plot(kind='bar', stacked=True)
```

```python
#From the above graph we infer GEORGIA has maximum number of complaints
```

```python
# Unresolved complaints distribution across State
df_unresolved = df[df['newStatus']=='Open']
colors = ['#639ace','#ca6b39','#7f67ca','#5ba85f','#c360aa','#a7993f','#cc566a']
df_unresolved = df_unresolved['State'].value_counts()
df_unresolved.head(25)
```

```python
df_unresolved.head().plot(kind='pie',autopct='%1.1f%%',
                        #explode = (0.15, 0, 0, 0, 0), startangle=45, shadow=False, colors = colors,
                        figsize = (4,3))
plt.axis('equal')
plt.title('# Unresolved complaints distribution across State\n')
plt.tight_layout()
plt.show()
```

```python
#From the above pie chart Georgia has maximum percentage of unresolved complaints
```

```python
#Provide the percentage of complaints resolved till date, which were received through the Internet and customer care calls
df_received = df[df['Received Via'].isin(['Internet','Customer Care Call'])]
df_received.head()
```

```python
df_received.newStatus.value_counts()
```

```python
df_received.newStatus.value_counts().plot(kind='pie',autopct='%1.1f%%',
                        #explode = (0.15, 0, 0, 0, 0), startangle=45, shadow=False, colors = colors,
                        figsize = (4,3))
plt.axis('equal')
plt.title('# complaints Status through Internet & Customer Care\n')
plt.tight_layout()
plt.show()
```
